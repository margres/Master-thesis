```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 2

```python
def plot_circle(r,x_0,y_0):
    '''
    function to plot a circle
    r: radius
    x_0, y_0 shift of the center
    '''
    
    theta = np.linspace(0, 2*np.pi, 100)
    x1 = r*np.cos(theta) + x_0
    x2 = r*np.sin(theta) + y_0
    return x1,x2
```

```python
def delta(beta,E=1):
    '''
    lens equation: gives image position given source position and Einstein radius
    '''
    return np.sqrt(beta**2+4*E**2)
```

```python

# the radius of the circle
r = 0.2
x_0=0.7
y_0=0.7

x_gal,y_gal=plot_circle(0.2,0.7,0.7)
x_lens,y_lens=plot_circle(1,0,0)

fig = plt.figure(dpi=150)
plt.plot(x_gal,y_gal, label='circular galaxy')
plt.plot(x_lens,y_lens, label='lens')

plt.gca().set_aspect('equal', adjustable='box')

plt.plot(0.5*(x_gal+delta(x_gal)),0.5*(y_gal+delta(y_gal)), label='image 1')
plt.plot(0.5*(x_gal-delta(x_gal)),0.5*(y_gal-delta(y_gal)), label='image 2')

plt.xlim(-2,2)
plt.ylim(-2,2)
params = {'legend.fontsize': 8,
          'legend.handlelength': 2}
plt.rcParams.update(params)
plt.legend(loc='lower right')
plt.show()
```

# Exercise 3

```python
def det_A(p): 
    x=p[0]
    y=p[1]
    C=(1+x**2+2*y**2)
    return 1+8/C**2 - 2*(3+2*x**2+2*y**2)*C**(-3/2)
```

```python
crit_1=np.array([0.766,0])
crit_2=np.array([3.873,0])
crit_3=np.array([0,0.871])
crit_4=np.array([0,1.225])
crit=np.concatenate((crit_1,crit_2,crit_3,crit_4,-crit_1,-crit_2,-crit_3,-crit_4)).reshape((8,2))
```

```python
det=[]
for c in crit:
    det.append(det_A(c))
```

```python
print('Determinant values for the given points:',det)
```

```python
def alpha(p):
    '''
    lens angle given the coordinates's of the point for the given deflection potential
    '''
    
    x=p[0]
    y=p[1]
    C=(1+x**2+2*y**2)
    n_1=2*x*C**(-1/2)
    n_2=4*y*C**(-1/2)
    
    return np.array([n_1,n_2])
```

```python
def beta(p):
    '''
    source plane coordinates mapped through the lens equation
    '''
    
    x=p[0]
    y=p[1]
    return p-alpha(p)
```

```python
point_caustic=np.zeros_like(crit)
for i,c in enumerate(crit):
    point_caustic[i,0]=beta(c)[0]
    point_caustic[i,1]=beta(c)[1]
    point_caustic[i,0]=beta(-c)[0]
    point_caustic[i,1]=beta(-c)[1]
    
print('Using the lens equation the intersection of the axes by the caustics are:',point_caustic)
```

```python
n_steps=800
xmin=-5
xmax=5

x_range=xmax-xmin
x_lin=np.linspace(xmin,xmax,n_steps)
y_lin=np.linspace(xmin,xmax,n_steps)

X,Y = np.meshgrid(x_lin, y_lin)

fig = plt.figure(dpi=100)

# contour
crit_curv = det_A((X,Y))
cp = plt.contour(X, Y, crit_curv,[0], colors='magenta',linestyles= '--', linewidths=0.7) 

#I get the coordinates of the contour plot
p_1 = cp.collections[0].get_paths()[0]  # grab the 1st path
coor_p_1 = p_1.vertices
x_crit_1 = coor_p_1[:,0]
y_crit_1 = coor_p_1[:,1]

p_2 = cp.collections[0].get_paths()[1]  # grab the 2nd path
coor_p2 = p_2.vertices
x_crit_2 = coor_p2[:,0]
y_crit_2 = coor_p2[:,1]

#I put the crtical curves' coordinates in the lens equation to get the caustics
caustic_1 = beta((x_crit_1,y_crit_1))
caustic_2 = beta((x_crit_2,y_crit_2))

plt.plot(caustic_1[0],caustic_1[1],caustic_2[0],caustic_2[1], c='b', lw=0.7)
plt.scatter(crit[:,0],crit[:,1], color='k', s=6,label='points critical curves')
plt.scatter(point_caustic[:,0],point_caustic[:,1], color='g', s=6,label='points caustics')

plt.gca().set_aspect('equal', adjustable='box')
cp.ax.set_ylabel('$\Theta_{y}$', fontsize=13)
cp.ax.set_xlabel('$\Theta_{x}$', fontsize=13)

plt.axhline(0,color='k',ls='--',linewidth=0.5)
plt.axvline(0,color='k',ls='--',linewidth=0.5)

plt.legend()
plt.show()
```
