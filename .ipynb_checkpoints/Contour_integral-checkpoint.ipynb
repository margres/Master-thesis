```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fft,ifft
from numpy import exp,arange
from scipy.optimize import curve_fit
from scipy import fftpack
from itertools import combinations
import scipy.optimize as op
from past.builtins import xrange
from scipy.optimize import fsolve
```

## Constants

```python
mu=0.1 
#shear in a 0-1 range
th=0.1
phi=0.1
```

## Constant step size

```python
def tau_point_mass(x_,y_): #time delay
    #tau=(1+z_l)/c*[d_os/(2*d_ol*d_ls)*(r-r_s)**2-2G/c**2]
    tau=(1+mu)*x_**2 +(1-mu)*y_**2 -2*th*(x_*np.cos(phi)+y_*np.sin(phi))-np.log(x_**2 +y_**2) #equation 4
    return tau

def polar_coord(r,theta, mu, th,phi): #the derivative if theta
    
    return r**2 * np.cos(theta)**2* (1+ mu) +(1-mu)*r**2*np.sin(theta)**2-2*th*r*(np.cos(theta)*np.cos(phi)+np.sin(theta)*np.sin(phi))- np.log(r**2) 
    
```

```python
n_steps=300
n_bins=300
xmin=-2.5
xmax=2.5
r_max= np.sqrt(2*xmax**2)
r=np.linspace(1*10^(-10),r_max, n_steps)
theta=np.linspace(0., 2*np.pi, n_steps)

#r,theta =np.meshgrid(r, theta) # grid of point
tau_polar= polar_coord(r,theta, mu, th,phi)

x_range=xmax-xmin
x_lin=np.linspace(xmin,xmax,n_steps)
y_lin=np.linspace(xmin,xmax,n_steps)

psi_t=[]
X,Y =np.meshgrid(x_lin, y_lin) # grid of point
tau = tau_point_mass(X, Y) # evaluation of the function on the grid

```

## find critical points with fsolve in x,y

```python
x_sol=[]
y_sol=[]


def equations(p):
    x,y=p
    der_x=2*x*(1+mu)-2*th*np.cos(phi)-2*x/(x**2+y**2)
    der_y=2*y*(1-mu)-2*th*np.sin(phi)-2*y/(x**2+y**2)
    return der_x,der_y

def dxx(x,y):
    return 2*(1+mu)-2*(x**2+y**2)/(x**2+y**2) +4*x**2/(x**2+y**2)**2

def dyy(x,y):
    return 2*(1-mu)-2/(x**2+y**2)+4*y**2/(x**2+y**2)**2

def dxy(x,y):
    return 4*x*y/(x**2+y**2)**2
    
def mu_critical(x,y): #i have to use the coordinates of the critical points, eq. 3.7 Nakamura 
    return 1/(dxx(x,y)*dyy(x,y)-dxy(x,y)**2)    
    
x_s=np.array((-1,1,-1,1))*0.5
y_s=np.array((-1,1,1,-1))*0.5


for x_,y_ in zip(x_s,y_s):
    x_tmp, y_tmp=fsolve(equations,(x_,y_)) #r,theta
    x_sol.append(x_tmp)
    y_sol.append(y_tmp)
```

## find critical points with fsolve in r,$\theta$ -- I know I have 4 images

```python
r_sol=[]
theta_sol=[]

def to_cartesian(r,theta):
    x=r*np.cos(theta)
    y=r*np.sin(theta)
    return x,y

def equations(p):
    r, theta = p
    #x,y=p
    cosTheta = np.cos(theta)
    sinTheta = np.sin(theta)
    der_theta=-2.*mu*np.sin(2.*theta)*r**2.-2.*th*r*(np.sin(phi)*cosTheta-np.cos(phi)*sinTheta)
    der_r=2.*r * cosTheta**2.* (1+ mu) +2.*r*(1-mu)*sinTheta**2.-2.*th*(cosTheta*np.cos(phi)+sinTheta*np.sin(phi))- 2./r
    return (der_r,der_theta)
    #return (2*(1+mu)*r*cosTheta-2*th*np.cos(phi)-2*cosTheta/r, 2*(1-mu)*r*sinTheta-2*th*np.sin(phi)-2*sinTheta/r)


node_theta_t = np.array([0,0.25*np.pi,0.5*np.pi,np.pi,1.5*np.pi,2.*np.pi])


for i in node_theta_t:
    r_tmp, theta_tmp=fsolve(equations,(1,i)) #r,theta
    if r_tmp>=0:
        r_sol.append(r_tmp)
        theta_sol.append(theta_tmp)
#print(equations((r, theta)))

#r_tmp, theta_tmp=fsolve(equations,(1,0)) #r,theta 

x=[]
y=[]
for t,r_ in zip(theta_sol,r_sol):
    x_tmp,y_tmp=to_cartesian(r_,t)
    x_tmp=round(x_tmp,6)
    y_tmp=round(y_tmp,6)
    if x_tmp not in x and y_tmp not in y:
        x.append(x_tmp)
        y.append(y_tmp)
        
```

```python
fig = plt.figure(dpi=600)
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes([left, bottom, width, height]) 
plt.scatter(x,y,color='r',s=4)
cp = ax.contour(X, Y, tau,np.linspace(0,2,50),linewidths=0.6, extent=[-2,2,-2,2], colors='black')

cp.ax.set_ylabel('y', fontsize=13)
cp.ax.set_xlabel('x', fontsize=13)

plt.title('Contour plot of time delay', fontsize=13)
#a=plt.colorbar(cset)
#a.set_label("time delay", fontsize=13)
plt.savefig('contour_plot_with_points')
plt.show()
```
