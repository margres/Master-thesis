```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fft,ifft
from numpy import exp,arange
```

## Constants

```python
mu=0.2 #shear in a 0-1 range
theta=0.1
phi=0.1
```

## Constant step size

```python
def tau_point_mass(x_,y_): #time delay
    #tau=(1+z_l)/c*[d_os/(2*d_ol*d_ls)*(r-r_s)**2-2G/c**2]
    tau=(1+mu)*x_**2 +(1-mu)*y_**2 -2*theta*(x_*np.cos(phi)+y_*np.sin(phi))-np.log(x_**2 +y_**2) #equation 4
    return tau
```

```python
n_steps=300
n_bins=300
x_lin=np.linspace(-2,2,n_steps)
y_lin=np.linspace(-2,2,n_steps)
psi_t=[]

'''
for a in x_lin:
    for b in y_lin:
        psi_t.append(tau_point_mass(a,b))
'''

X,Y =np.meshgrid(x_lin, y_lin) # grid of point
tau = tau_point_mass(X, Y) # evaluation of the function on the grid

#dx=np.ones(len(tau))*4/n_steps
```

```python
tau_for_hist=tau.flatten()
dx=np.ones(len(tau_for_hist))*4/(n_steps)
```

### Histogram

```python
values, bins, _ =plt.hist(tau_for_hist,n_bins, weights=dx*dx/(np.max(tau)-np.min(tau))*n_bins) #psi
#values, bins, _ =plt.hist(tau_for_hist,n_bins, weights=dx*dx*(np.max(tau)-np.min(tau))/n_bins) #psi
plt.xlim(0.5, 7)
area = sum(np.diff(bins)*values)
#plt.xlim(0, 1)
plt.title('Bins '+str(n_bins)+' Steps '+str(n_steps))
plt.show()
plt.savefig('hist_constantstep_Bins_'+str(n_bins)+'_Steps_'+str(n_steps))
```

```python
x=-2
y=-2
t=[]
stepsize=[]
threshold=0.1
constant_step_size=1/400
dx=0

for i in range(30000):
    #tau.append(tau_point_mass(x,y))

    for_now= tau_point_mass(x+dx,y+dx)
    
    if i==0 or abs(for_now-t[i-1]) <threshold :
        #print(i)
        #for_now= tau_point_mass(x,y)
        t.append(for_now)    
        dx=constant_step_size #i go back to the usual step size 


    
    else:   #it means it is changing fast, i.e. where contour lines are denser
        while abs(for_now-t[i-1])> threshold:
            dx=dx-1/(400*10)  #i can divide in half everytime also
         
            if dx<0 or dx<0.00001:
                dx=dx+1/(400*10)
                #print(dx)
                break 
                
            for_now=tau_point_mass(x+dx,y+dx)
        t.append(for_now)
    if t[i]<0.2:
        print(x, t[i])
    x=x+dx
    y=y+dx
    stepsize.append(dx) #keep track of the weights
    if (x+dx)>2:
        print('done')
        break



  
        
```

```python
def polar_coord(r,d):
    return r**2 * np.cos(d)**2* (1+ mu) +(1-mu)*r**2*np.sin(d)**2-2*theta*r*(np.cos(d)*np.cos(phi)+np.sin(d)*np.sin(phi))- np.log(r**2) 

time=np.linspace(0,0.5,n_steps)
d= np.linspace(0,180,n_steps)

def b():
    return 2*(1+theta*(np.cos(d)*np.cos(phi)+np.sin(d)*np.sin(phi)))

def a():
    return mu*np.cos(2*d)
    
def c(constant_time_delay):
    return -1-constant_time_delay

def sol(constant_time_delay):
    sol_plus=(-b()+np.sqrt(b()**2-4*a()*c(constant_time_delay)))/2
    sol_min= (-b()-np.sqrt(b()**2-4*a()*c(constant_time_delay)))/2
    return sol_plus, sol_min
              
```

```python
time=np.linspace(0,2,n_steps)
solutions_min=[]
solutions_plus=[]
for time_ in time:
    #print(time_)
    solutions_plus.append(sol(time_)[0])
    solutions_min.append(sol(time_)[1])
```
