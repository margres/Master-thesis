```python
def dift(x):
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(2j * np.pi * k * n / N)
    return np.dot(M, x)
```

```python
plt.plot(omega,np.abs(dift(ys)))
```

```python
def FT_crit(omega,T,mu, xs, ys):
    '''
    analytical contribution of saddle points

    '''
    
    if mu<0:
        return 2j*np.exp(1j*omega*T)*mu
    else:
        return -2*np.exp(1j*omega*T)
def magnification(F):
    return np.abs(F)**2
```

```python
for i,(m,t) in enumerate(zip(muI,tauI)):
        F_crit[i,:]=FT_crit(omega,t,m, xs, ys)
        
```

```python
F_crit=np.sum(F_crit,axis=0)
plt.plot(omega,magnification(F_crit))
plt.save()
```
